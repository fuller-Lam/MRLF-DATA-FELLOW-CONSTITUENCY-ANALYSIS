```python
benchmark_metrics = [
    ("Electorate_2023", "Electorate"),
    ("Literacy_Rate_Pct", "Literacy rate %"),
    ("MGNREGS_Person_Days_FY2425", "MGNREGS person-days"),
    ("PM_Awas_Completion_Pct", "PM Awas completion %"),
    ("JJM_Functional_Tap_Connections_Pct", "JJM tap coverage %"),
    ("PM_KISAN_Beneficiaries", "PM-KISAN beneficiaries"),
    ("Road_Length_Km", "Road length km"),
    ("Pucca_Road_Pct", "Pucca road %"),
    ("Internet_4G_Coverage_Pct", "4G coverage %"),
    ("Health_Sub_Centres", "Health sub-centres"),
    ("Primary_Health_Centres", "Primary health centres"),
    ("Community_Health_Centres", "Community health centres"),
    ("Govt_Primary_Schools", "Government primary schools"),
    ("Govt_Secondary_Schools", "Government secondary schools"),
]

def compare_to_average(value_in, average_in, tolerance=1e-9):
    if pd.isna(value_in) or pd.isna(average_in):
        return "- unavailable"
    if abs(value_in - average_in) <= tolerance:
        return "→ at avg"
    if value_in > average_in:
        return "↑ above avg"
    return "↓ below avg"

def format_number(value_in, decimals=1):
    if pd.isna(value_in):
        return "N/A"
    if float(value_in).is_integer():
        return format(int(value_in), ",")
    return format(round(float(value_in), decimals), ",")

def build_profile(df_in, constituency_name):
    match_df = df_in[
        df_in["Constituency_Name"].astype(str).str.strip().str.lower() == constituency_name.strip().lower()
    ]

    if match_df.empty:
        raise ValueError("Constituency name not found: " + str(constituency_name))

    row = match_df.iloc[0]
    region_df = df_in[df_in["Region"] == row["Region"]]
    state_avg = df_in.mean(numeric_only=True)
    region_avg = region_df.mean(numeric_only=True)

    total_health = int(
        row["Health_Sub_Centres"] +
        row["Primary_Health_Centres"] +
        row["Community_Health_Centres"]
    )

    total_schools = int(
        row["Govt_Primary_Schools"] +
        row["Govt_Secondary_Schools"]
    )

    lines = []
    lines.append("=" * 72)
    lines.append("CONSTITUENCY PROFILE: " + str(row["Constituency_Name"]) + " (" + str(row["Constituency_ID"]) + ")")
    lines.append("=" * 72)
    lines.append("")
    lines.append("Constituency snapshot")
    lines.append("- Region: " + str(row["Region"]))
    lines.append("- District: " + str(row["District"]))
    lines.append("- Electorate: " + format_number(row["Electorate_2023"], 0))
    lines.append("- Primary occupation: " + str(row["Primary_Occupation"]))
    lines.append("- Literacy rate: " + format_number(row["Literacy_Rate_Pct"]) + "%")
    lines.append("")
    lines.append("Scheme status")
    lines.append("- MGNREGS person-days: " + format_number(row["MGNREGS_Person_Days_FY2425"], 0))
    lines.append("- PM Awas completion: " + format_number(row["PM_Awas_Completion_Pct"]) + "%")
    lines.append("- JJM tap coverage: " + format_number(row["JJM_Functional_Tap_Connections_Pct"]) + "%")
    lines.append("- PM-KISAN beneficiaries: " + format_number(row["PM_KISAN_Beneficiaries"], 0))
    lines.append("")
    lines.append("Infrastructure")
    lines.append("- Road length: " + format_number(row["Road_Length_Km"]) + " km")
    lines.append("- Pucca road coverage: " + format_number(row["Pucca_Road_Pct"]) + "%")
    lines.append("- 4G coverage: " + format_number(row["Internet_4G_Coverage_Pct"]) + "%")
    lines.append(
        "- Health facilities: " + str(total_health) + " total (" +
        str(int(row["Health_Sub_Centres"])) + " sub-centres, " +
        str(int(row["Primary_Health_Centres"])) + " PHCs, " +
        str(int(row["Community_Health_Centres"])) + " CHCs)"
    )
    lines.append(
        "- Schools: " + str(total_schools) + " total (" +
        str(int(row["Govt_Primary_Schools"])) + " primary, " +
        str(int(row["Govt_Secondary_Schools"])) + " secondary)"
    )
    lines.append("")
    lines.append("Benchmarking")
    for metric_key, metric_label in benchmark_metrics:
        value_text = format_number(row[metric_key])
        state_flag = compare_to_average(row[metric_key], state_avg[metric_key])
        region_flag = compare_to_average(row[metric_key], region_avg[metric_key])
        lines.append(
            "- " + metric_label + ": " + value_text +
            " | State: " + state_flag +
            " | Region: " + region_flag
        )
    lines.append("")
    lines.append("Fund utilisation")
    fund_util = row["Constituency_Fund_Utilized_Pct"]
    lines.append("- Allocated: " + format_number(row["Constituency_Fund_Allocated_Lakhs"]) + " lakhs")
    lines.append("- Utilized: " + format_number(fund_util) + "%")

    if fund_util < 60:
        lines.append("! WARNING: Fund utilisation is below 60% and requires attention.")

    return "\n".join(lines)
```

```python
target_constituencies = ["Mawlai", "Pynthorumkhrah", "Nongthymmai"]

for constituency_name in target_constituencies:
    profile_text = build_profile(df_constituencies, constituency_name)
    print(profile_text)
    print("\n")
```
