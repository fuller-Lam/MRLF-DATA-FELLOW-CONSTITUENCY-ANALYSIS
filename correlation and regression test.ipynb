```python
import pandas as pd
import numpy as np

df = pd.read_csv("C:/Users/ohabi/OneDrive/Desktop/MRLF DATA FELLOW ASSIGNMENTS/corelation plots/DATA FOR CORRELATION TEST QUERY 2.csv")

df.head()
```

```python
df['BPL_Rank'] = df['BPL_Households_Pct'].rank(
    ascending=False,
    method='dense'
)

df['JJM_Rank'] = df['JJM_Functional_Tap_Connections_Pct'].rank(
    ascending=False,
    method='dense'
)

df['PMAY_Rank'] = df['PM_Awas_Completion_Pct'].rank(
    ascending=False,
    method='dense'
)
```

```python
df['JJM_Rank_Gap'] = df['BPL_Rank'] - df['JJM_Rank']

df['PMAY_Rank_Gap'] = df['BPL_Rank'] - df['PMAY_Rank']
```

```python
jjm_top = df.sort_values(
    'JJM_Functional_Tap_Connections_Pct',
    ascending=False
)[[
    'Constituency_Name',
    'JJM_Functional_Tap_Connections_Pct',
    'JJM_Rank'
]]

jjm_top.head(10)
```

```python
pmay_top = df.sort_values(
    'PM_Awas_Completion_Pct',
    ascending=False
)[[
    'Constituency_Name',
    'PM_Awas_Completion_Pct',
    'PMAY_Rank'
]]

pmay_top.head(10)
```

```python
from scipy.stats import spearmanr
```

```python
corr_jjm, p_jjm = spearmanr(
    df['BPL_Households_Pct'],
    df['JJM_Functional_Tap_Connections_Pct']
)

print("JJM")
print("Correlation =", round(corr_jjm,3))
print("P-value =", round(p_jjm,4))
```

```python
corr_pmay, p_pmay = spearmanr(
    df['BPL_Households_Pct'],
    df['PM_Awas_Completion_Pct']
)

print("PMAY")
print("Correlation =", round(corr_pmay,3))
print("P-value =", round(p_pmay,4))
```

```python
correlation_results = pd.DataFrame({
    'Relationship':[
        'BPL vs JJM Coverage',
        'BPL vs PMAY Completion'
    ],
    'Spearman_r':[
        corr_jjm,
        corr_pmay
    ],
    'P_Value':[
        p_jjm,
        p_pmay
    ]
})

correlation_results
```

```python
import statsmodels.formula.api as smf
```

```python
pmay_model = smf.ols(
    'PM_Awas_Completion_Pct ~ BPL_Households_Pct',
    data=df
).fit()

print(pmay_model.summary())
```

```python
jjm_model = smf.ols(
    'JJM_Functional_Tap_Connections_Pct ~ BPL_Households_Pct',
    data=df
).fit()

print(jjm_model.summary())
```

```python
jjm_regression = pd.DataFrame({
    'Variable': jjm_model.params.index,
    'Coefficient': jjm_model.params.values,
    'P_Value': jjm_model.pvalues.values,
    'Std_Error': jjm_model.bse.values
})
```

```python
pmay_regression = pd.DataFrame({
    'Variable': pmay_model.params.index,
    'Coefficient': pmay_model.params.values,
    'P_Value': pmay_model.pvalues.values,
    'Std_Error': pmay_model.bse.values
})
```

```python
with pd.ExcelWriter(
    "Scheme_Delivery_Analysis.xlsx",
    engine="openpyxl"
) as writer:

    df.to_excel(
        writer,
        sheet_name="Constituency_Ranks",
        index=False
    )

    correlation_results.to_excel(
        writer,
        sheet_name="Correlations",
        index=False
    )

    jjm_regression.to_excel(
        writer,
        sheet_name="JJM_Regression",
        index=False
    )

    pmay_regression.to_excel(
        writer,
        sheet_name="PMAY_Regression",
        index=False
    )

print("Export complete.")
```
